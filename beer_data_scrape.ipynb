```python
html_page2 = requests.get('https://www.beeradvocate.com/place/reviews/')
soup_recent = BeautifulSoup(html_page2.content, 'html.parser')
```

```python
soup_recent.select('a[href*=/community/members/]')[7].text
```

```python
user_scrape = soup_recent.select('a[href*=/community/members/]')
```

```python
#get user names from the recent reviews page
def users():
    user_list = []
    for x in user_scrape:
        if x.text != '':
            user_list.append(x.text)
    return list(set(user_list))
recent_users = users()
```

```python
html_page3 = requests.get('https://www.beeradvocate.com/lists/top/')
soup_top = BeautifulSoup(html_page3.content, 'html.parser')
```

```python
#urls for top 250 beers
def top_beer_url():
    beer_url_list = []
    for x in soup_top.select('a[href*=/beer/profile/]'):
        beer_url_list.append('https://www.beeradvocate.com' + x.get('href'))
    return beer_url_list[0::2]
beer_url_list = top_beer_url()
```

```python
beer_url_list[0]
```

```python
#top 250 beers, the 25 most recent reviewers
def user_top_scrape():
    top_scrape_list = []
    for x in beer_url_list:
        html_page = requests.get(x)
        soup_top_scrape = BeautifulSoup(html_page.content, 'html.parser')
        user_top_scrape = soup_top_scrape.select('a[href*=/community/members/]')
        for y in user_top_scrape:
            if y.text != '':
                top_scrape_list.append(y.text)
    return list(set(top_scrape_list))
top_beer_users = user_top_scrape()
```

```python
def user_beer_info():
    total_list = []
    for x in user_profiles:
        html_page = requests.get(x)
        bf_scrape = BeautifulSoup(html_page.content, 'html.parser')
        beer_name_brew = bf_scrape.select('a[href*=/beer/profile]')
        beer_type = bf_scrape.select('a[href*=/beer/styles/]')[2:]
        user_score = bf_scrape.select('td[valign]')[5::7]
        beer_abv = bf_scrape.select('td[valign]')[4::7]
        nb_list = []
        type_list = []
        score = []
        abv = []
        for x in beer_name_brew:
            nb_list.append(x.text)
        for x in beer_type:
            type_list.append(x.text)
        for x in user_score:
            score.append(x.text)
        for x in beer_abv:
            abv.append(x.text)
        name_text = nb_list[0::2]
        brew_text = nb_list[1::2]
        type_text = type_list
        score_text = score
        abv_text = abv
        total_list.append(list(zip(name_text, brew_text, type_text, score_text, abv_text)))
    return total_list
final_beer_info = user_beer_info()
```
